```
from pycaret.datasets import get_data
dataset = get_data('jewellery')
```

```
type(dataset)
```

```
dataset.head()
```

```
from pycaret.clustering import *
reg = setup(dataset, normalize=True, silent=True)
```

```
kmeans = create_model('kmeans', num_clusters=5)
```

```
predict = assign_model(kmeans)
```

```
predict.head()
```

```
p = predict_model(model=kmeans, data=dataset)
```

```
p.head()
```

```
tuned_kmeans = tune_model(kmeans, supervised_target='Age')
```
